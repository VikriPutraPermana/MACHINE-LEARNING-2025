```python
# Import library
import pandas as pd

# Ganti path sesuai lokasi file kamu
path = '../data/day.csv'

# Baca dataset
df = pd.read_csv(path)

# Tampilkan hasil
print("✅ Dataset berhasil dibaca!")
print(f"Jumlah total data: {len(df)}")
display(df.head())
```

```python
# Import modul pembagi data
from sklearn.model_selection import train_test_split

# Pisahkan data Training (80%) dan Testing (20%)
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Tampilkan hasil
print("✅ Data Training berhasil dibuat!")
print(f"Jumlah Data Training: {len(train_data)}")
display(train_data.head())
```

```python
# Pisahkan 10% dari data training untuk validation
train_data, val_data = train_test_split(train_data, test_size=0.1, random_state=42)

# Tampilkan hasil
print("✅ Data Validation berhasil dibuat!")
print(f"Jumlah Data Validation: {len(val_data)}")
display(val_data.head())
```

```python
# Data testing sudah otomatis terbentuk dari langkah pertama
print("✅ Data Testing berhasil dibuat!")
print(f"Jumlah Data Testing: {len(test_data)}")
display(test_data.head())
```

```python
print("=== REKAP JUMLAH DATA ===")
print(f"Total Data Asli   : {len(df)}")
print(f"Data Training     : {len(train_data)}")
print(f"Data Validation   : {len(val_data)}")
print(f"Data Testing      : {len(test_data)}")
```
